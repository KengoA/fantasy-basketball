```python
import glob
import pandas as pd
import seaborn as sns
from matplotlib     import pyplot as plt
%matplotlib inline
```

```python
#Concatenate all csv files under a directory
def csv_concatenate(folder_path):
    files = glob.glob(folder_path + "/*.csv")
    df_list = []
    for file in files:
        df_list.append(pd.read_pickle(file))
    df = pd.concat(df_list)
    return df
```

```python
df = csv_concatenate('/home/kengo/NBA/data/2016-2017/').reset_index(drop=True)
```

```python
pd.set_option("display.max_columns", 27)
df_new = df.sort_values(by='3P',ascending=False)
```

```python
sns.set_style('dark')
sns.lmplot(x='FG',y='FG_perc',data=df,size=10)
```

```python
names = ['Kyrie Irving','Kyle Korver','LeBron James','Stephen Curry','Kevin Durant','Isaiah Thomas',
         'James Harden','Russell Westbrook','Malcolm Brogdon','Jonas Valanciunas']

for name in names:
    dict_player = {'Name':[],'Date':[],'PTS':[],'AST':[],'FG':[],'FG_perc':[],'3P':[],'3P_perc':[]}
    for i in range(len(df)):
        if df['Name'][i] == name and df['MP'][i] >= 10.00 and df['FGA'][i] >= 3 and df['3PA'][i] >= 3:
            dict_player['Name'].append(name)
            dict_player['Date'].append(pd.to_datetime(df['Date'][i],format='%Y%m%d'))
            dict_player['PTS'].append(df['PTS'][i])
            dict_player['AST'].append(df['AST'][i])
            dict_player['FG'].append(df['FG'][i])
            dict_player['3P'].append(df['3P'][i])
            dict_player['FG_perc'].append(df['FG_perc'][i])
            dict_player['3P_perc'].append(df['3P_perc'][i])
    df_player = pd.DataFrame(dict_player).loc[:,['Name','Date','PTS','AST','FG','FG_perc','3P','3P_perc']]
    df_player.plot(x='Date',y=['FG_perc','3P_perc'],figsize=(20,10),title=df_player['Name'][0],kind='line')
    df_player.plot(x='Date',y=['FG','3P'],figsize=(20,10),title=df_player['Name'][0],kind='bar')
```

```python
df_player = df_player.sort_values(by='PTS',ascending=False)
```
